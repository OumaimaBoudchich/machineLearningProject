```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
df = pd.read_csv("installments_payments.csv")
dfjoin = pd.read_csv("application_train.csv")
df = df.merge(dfjoin[["SK_ID_CURR","TARGET"]],how="left",on="SK_ID_CURR")
```

```python
df=df.dropna()
df.describe()
```

```python
df.corr()
```

```python
dfg=df.groupby("SK_ID_CURR",as_index=False).agg({"DAYS_INSTALMENT":"mean",
                            "DAYS_ENTRY_PAYMENT":"mean","TARGET":"min"})
```

```python
dfg[["SK_ID_CURR","DAYS_INSTALMENT","DAYS_ENTRY_PAYMENT"]]
```

```python
dfg.to_csv("installment.csv")
```
